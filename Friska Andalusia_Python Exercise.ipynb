https://www.w3resource.com/python-exercises/string/python-data-type-string-exercise-2.php


Tulis program Python untuk membuat kamus dari string. Pergi ke editor

__Catatan:__  Hitung jumlah huruf dari `string`

Contoh `string`: 'G-Incubation'

Keluaran yang diharapkan:
```
{'G': 1, '-': 1, 'I': 1, 'n': 2, 'c': 1, 'u': 1, 'b': 1, 'a': 1, 't': 1, 'i': 1, 'o': 1}
```

````python
def frekuensi(str1):
    dict ={}
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] = +1
        else:
            dict[n]=1
    return dict
````

````python
print(frekuensi('G-Incubation'))
````

###  Tugas 3 : Perbandingan `String`

#### Skenario Kasus
Buat sebuah program berupa permainan gunting batu kertas yang dapat dimainkan oleh dua orang. Kedua pemain menginput pilihan masing-masing __(gunting/batu/kertas)__, kemudian komputer melakukan perbandingan untuk melihat dan menampilkan siapa yang menang.


> (catatan: untuk menyembunyikan input dari user, dapat digunakan fungsi *getpass()* dan untuk input user gunakkan fungsi *input()* )


````python
import getpass

def batu_gunting_kertas():
    print('Pilih Batu, Kertas, Gunting!')
    mulai_main = True
    while mulai_main:
        while True:
            validinput =['gunting','batu','kertas']
            
            p1_input = getpass.getpass('P1: Batu,Kertas Or Gunting?')
            
            while p1_input.lower() not in validinput:
                p1_input =getpass.getpass ('P1: Batu,Kertas Or Kertas aja:')
                
            p2_input = getpass.getpass('P2: Batu,Kertas Or Gunting?')
            
            while p2_input.lower() not in validinput:
                p2_input =getpass.getpass ('P2: Batu,Kertas Or Kertas aja:')
                
            player_inputs = [p1_input.lower(), p2_input.lower()]
            
            kertas_menang = ['batu','kertas']
            batu_menang = ['batu','gunting']
            gunting_menang = ['gunting','kertas']
            
            if player_inputs == batu_menang or player_inputs == batu_menang[::-1]:
                player_name = 'P1' if p1_input == 'batu' else 'P2'
                print(player_name + ' menang')
                break
                
            elif player_inputs == kertas_menang or player_inputs == kertas_menang[::-1]:
                player_name = 'P1' if p1_input == 'kertas' else 'P2'
                print(player_name + ' menang')
                break
                
            elif player_inputs == gunting_menang or player_inputs == gunting_menang[::-1]:
                player_name = 'P1' if p1_input == 'gunting' else 'P2'
                print(player_name + ' menang')
                break
                
            else:
                print('seri')
                break
                
        play_again = input('Ulang Permainan (Y/N)')
        valid_mulai_main_lagi = ['Y','N']
        while play_again.upper() not in valid_mulai_main_lagi:
            play_again = input ('Pilih Y atau N:')
        mulai_main = True if play_again.upper() == 'Y' else False
        
batu_gunting_kertas()
                
            
````

###  Tugas 4 : Mengukur Berat Badan
https://koding.alza.web.id/latihan-soal-percabangan/
##### Skenario Kasus
BMI (Body Mass Index) adalah salah satu cara untuk menentukan apakah seseorang bertubuh gemuk, langsing, atau kurus berdasarkan hasil bagi massa tubuh (kilogram) dengan tinggi badan (meter) kuadrat.

Nilai BMI di bawah 18.5 berarti terlalu kurus, langsing/sehat pada rentang 18.5 s.d di bawah 25, 25+ tergolong gemuk.

Buatlah sebuah program yang menanyakan berapa berat badan seseoarang (dalam kg), dan berapa tingginya (dalam cm, 1m = 100 cm), kemudian hitung nilai BMInya. Setelah nilai BMI didapatkan, cetak nilai BMI dan apakah orang tersebut termasuk kurus, langsing, atau gemuk. Lihat contoh output berikut untuk lebih jelasnya.

##### Hasil Keluaran :
```
Masukkan berat badan anda (kg): 65
Masukkan tinggi badan anda (cm): 172
Nilai BMI anda adalah 21,9713
Anda termasuk berbadan langsing
```

````python
bb = input('Masukkan berat badan anda (kg): ')
tb = input('Masukkan tinggi badan anda (cm): ')
bb = float(bb)
tb = float(tb) / 100

bmi = bb / (tb * tb)
print('Nilai BMI anda adalah: ',bmi)
if bmi < 18.5:
  print('Anda tergolong berbadan kurus')
elif bmi < 25:
  print('Anda tergolong berbadan langsing')
else:
  print('Anda tergolong berbadan gemuk')

````

Tulis program Python untuk membuat pola berikut, menggunakan loop for bersarang.

         *

         *  *  *  * 
         *  *  *  *  * 
         *  *  *  *  *  * 
         *  *  *  *  *  *  * 
         *  *  *  *  *  *  *  * 
         *  *  *  *  *  *  *  *  * 
         *  *  *  *  *  *  *  *  *  * 


````python
string = ""
bar = 1

x = int(input("Masukkan angka :"))

# Looping Baris
while bar <= x:
    kol = bar

    # Looping Kolom
    while kol > 0:
        string = string + " * "
        kol = kol - 1

    string = string + "\n"
    bar = bar + 1
print (string)
````

### Tugas 4 : Program Bintang 2

https://codesaya.com/a/tutorial-membuat-segitiga-dengan-python-versi-bintang-mxzpdhnhux/

Tulis program Python untuk membuat pola berikut, menggunakan loop for bersarang.


```

 *  *  *  *  *  *  *  *  *  *  *  *  *  *  *  *  *  *  * 

    *  *  *  *  *  *  *  *  *  *  *  *  *  *  *  *  * 

       *  *  *  *  *  *  *  *  *  *  *  *  *  *  * 

          *  *  *  *  *  *  *  *  *  *  *  *  * 

             *  *  *  *  *  *  *  *  *  *  * 

                *  *  *  *  *  *  *  *  * 

                   *  *  *  *  *  *  * 

                      *  *  *  *  * 

                         *  *  * 

                            * 
```


````python
string = ""
bar = 1

x = int(input("Masukkan angka :"))
print ("\n")
# Looping Baris
while bar <= x:
    kol = bar
# Looping Kolom Spasi Kosong
    while kol > 1:
        string = string + "   "
        kol = kol - 1
# Looping Kolom Bintang Sisi Kiri	
    kiri = 0
    while kiri <= (x - bar):
        string = string + " * "
        kiri = kiri + 1	
# Looping Kolom Bintang Sisi Kanan
    kanan = kiri
    while kanan > 1:
        string = string + " * "
        kanan = kanan - 1

    string = string + "\n\n"
    bar = bar + 1
print (string)
````

### Tugas 5 : Program Menghitung Huruf Vokal
https://jagongoding.com/python/latihan-logika/menghitung-huruf-vokal/#skenario-kasus

#### Skenario Kasus
Alur programnya;

1. Program  akan minta user untuk memasukkan sebuah teks
2. Kemudian program akan menampilkan jumlah huruf vokal pada teks tersebut

#### Contoh Keluaran
```
Tuliskan teks; Indonesia Adalah Negeri Sejuta Pulau
Total karakter: 36
Total huruf vokal: 17
  a -> 6
  i -> 3
  u -> 3
  e -> 4
  o -> 1
 ```
  


````python
teks = input('Tuliskan teks: ').lower()
hrf_vokal = {
  'a': 0,
  'i': 0,
  'u': 0,
  'e': 0,
  'o': 0
}

total_h_vokal = 0

for karakter in teks:
  if karakter in ['a', 'i', 'u', 'e',  'o']:
    hrf_vokal[karakter] += 1
    total_h_vokal += 1
    
print(f'Total karakter: {len(teks)}')
print(f'Total huruf vokal: {total_h_vokal}')
print(f"""\
  a -> {hrf_vokal['a']}
  i -> {hrf_vokal['i']}
  u -> {hrf_vokal['u']}
  e -> {hrf_vokal['e']}
  o -> {hrf_vokal['o']}
\
""")
````
